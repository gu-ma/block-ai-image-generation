## Step 1 — Imports and setup

```python
import os
import torch
from diffusers import DiffusionPipeline

device = 'cuda' if torch.cuda.is_available() else 'cpu'
dtype = torch.float16 if device == 'cuda' else torch.float32
os.makedirs('generated_outputs', exist_ok=True)

print('Device:', device)
print('DType:', dtype)
```

## Step 2 — Load pipeline (Quickstart pattern)

```python
model_id = 'stabilityai/sdxl-turbo'

if device == 'cuda':
    pipeline = DiffusionPipeline.from_pretrained(
        model_id,
        torch_dtype=torch.float16,
        variant='fp16'
    )
else:
    pipeline = DiffusionPipeline.from_pretrained(
        model_id,
        torch_dtype=torch.float32
    )

pipeline = pipeline.to(device)
print('Loaded:', model_id)
```

## Step 3 — First generation
For Turbo, start with `num_inference_steps=1` and `guidance_scale=0.0`.

```python
prompt = 'A cute robot reading a book in a cozy library, cinematic lighting'
negative_prompt = 'blurry, low quality, deformed'
seed = 42

generator = torch.Generator(device=device).manual_seed(seed)
image = pipeline(
    prompt=prompt,
    negative_prompt=negative_prompt,
    num_inference_steps=1,
    guidance_scale=0.0,
    generator=generator,
).images[0]

image.save('generated_outputs/first_image.png')
image
```

## Step 4 — Tiny helper for experiments

```python
def generate(prompt_text, seed=42, steps=1, guidance=0.0, filename='out.png'):
    gen = torch.Generator(device=device).manual_seed(seed)
    img = pipeline(
        prompt=prompt_text,
        negative_prompt=negative_prompt,
        num_inference_steps=steps,
        guidance_scale=guidance,
        generator=gen,
    ).images[0]
    path = f'generated_outputs/{filename}'
    img.save(path)
    return img, path
```

## Experiment A — Change prompt only

```python
img_a1, _ = generate('A cute robot reading a book in a cozy library', seed=42, steps=1, guidance=0.0, filename='a_prompt_1.png')
img_a2, _ = generate('A cute robot reading a book in a futuristic neon library, ultra detailed', seed=42, steps=1, guidance=0.0, filename='a_prompt_2.png')
display(img_a1)
display(img_a2)
```

## Experiment B — Change seed only

```python
img_b1, _ = generate(prompt, seed=7, steps=1, guidance=0.0, filename='b_seed_7.png')
img_b2, _ = generate(prompt, seed=8, steps=1, guidance=0.0, filename='b_seed_8.png')
display(img_b1)
display(img_b2)
```

## Experiment C (optional) — Change steps and guidance

```python
img_c, p_c = generate(prompt, seed=42, steps=2, guidance=1.0, filename='c_steps2_guidance1.png')
display(img_c)
print(p_c)
```
